```python
from lib.Locate_ppl import Locate_ppl
from lib.PersonEmbeding import PersonEmbeding
from lib.EmbedingsManiger import EmbedingsManiger
from matplotlib import pyplot as plt
import torch
```

```python
lp = Locate_ppl()
pe = PersonEmbeding()
emb_maniger = EmbedingsManiger(cache_ram=True)
```

```python
def get_data(data, max_it=3):
    c = 0
    for img in lp:
        if len(img)==0:
            continue
        print(c)
        emb = pe.gen_training_emb(img)
        for e in emb:
            data.append(e)
        c+=1
        if c >= max_it:
            break
    return img[0]
```

```python
lp.from_disk = True
for path, name in [("images/kiri", "Kiri"), ("images/samuel", "Samuel")]:
    print(name)
    if name in emb_maniger.info:
        emb_maniger.remove(name)
    lp.set_path(path)
    data = []
    img = get_data(data, 1000)
    emb_maniger.new(name, img, data)
```
